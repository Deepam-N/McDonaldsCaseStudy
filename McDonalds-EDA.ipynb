```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
from sklearn.preprocessing import LabelEncoder
```

```python
# Load the dataset
data = pd.read_csv('mcdonalds.csv')
```

```python
# Display basic information about the dataset
print(data.info())
```

```python
# Display the first few rows of the dataset
print(data.head())
```

```python
# Summary statistics
print(data.describe())
```

```python
# Check for missing values
print(data.isnull().sum())
```

```python
# Distribution of age
plt.figure(figsize=(8, 6))
sns.histplot(data['Age'], bins=20, kde=True)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Gender distribution
plt.figure(figsize=(6, 6))
sns.countplot(data=data, x='Gender')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
# Visualize perception attributes
perception_attributes = data.columns[:-4]  # Exclude 'Like', 'Age', 'VisitFrequency', and 'Gender'
plt.figure(figsize=(12, 10))
for i, attr in enumerate(perception_attributes, 1):
    plt.subplot(4, 3, i)
    sns.histplot(data[attr], bins=5, kde=True)
    plt.title(attr)
    plt.xlabel('Rating')
    plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
# Relationship between age and liking
plt.figure(figsize=(8, 6))
sns.scatterplot(data=data, x='Age', y='Like')
plt.title('Age vs Liking')
plt.xlabel('Age')
plt.ylabel('Like')
plt.show()
```

```python
label_encoder = LabelEncoder()
data['VisitFrequency'] = label_encoder.fit_transform(data['VisitFrequency'])
```

```python
# Relationship between visit frequency and liking
plt.figure(figsize=(8, 6))
sns.boxplot(data=data, x='VisitFrequency', y='Like')
plt.title('Visit Frequency vs Liking')
plt.xlabel('Visit Frequency')
plt.ylabel('Like')
plt.xticks(rotation=45)
plt.show()
```
